```python
# Questão 07
import re

def contadorDeEspacosNaoSimples(arquivo):
    with open(arquivo, 'r', encoding='utf-8') as arquivoLido:
        for linha in arquivoLido:
            regex = re.sub('[ ]+', ' ', linha)
            with open('resultadoQuestão07.txt', 'a', encoding='utf-8') as arquivoEscrito:
                arquivoEscrito.write(regex)
    return 'resultadoQuestão07.txt'

contadorDeEspacosNaoSimples('Operadores.html')
```

```python
#Questão 08 
import re

def extrairNumeros(arquivo):
    dictNum = {}
    with open(arquivo, 'r', encoding='utf-8') as arquivoLido:
        for linha in arquivoLido:
            regex = re.findall('[0-9]', linha)
            for item in regex:
                if item in dictNum:
                    dictNum[item] += 1
                else:
                    dictNum[item] = 1
    return dictNum
    
extrairNumeros('Operadores.html')
```

```python
#Questão 09
import re

def encontrarMaiorMenorPalavra(nomeArquivo):
    listaMenores, listaMaiores = [], []
    with open(nomeArquivo, 'r', encoding='utf-8') as arquivo:
        arquivoLido = arquivo.read()
        separarPalvras = re.findall('[^\d\W]+', arquivoLido)
        menor = separarPalvras[0]
        for palavra in separarPalvras:
            if len(palavra) < len(menor):
                menor = palavra
        for palavra in separarPalvras:
            if len(palavra) == len(menor):
                listaMenores.append(palavra)

        maior= separarPalvras[0]
        for palavra in separarPalvras:
            if len(palavra) > len(maior):
                maior = palavra
        for palavra in separarPalvras:
            if len(palavra) == len(maior):
                listaMaiores.append(palavra)
    return listaMaiores, listaMenores

encontrarMaiorMenorPalavra('Operadores.html')
```
